```python
import pandas as pd
inmuebles = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/inmersion_de_datos/inmuebles_bogota.csv')
columnsa = {'Baños':'Banos', 'Área':'Area'}
inmuebles = inmuebles.rename(columns=columnsa)
inmuebles.head()
```

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)
```

```python
inmuebles.info()
```

```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

three_inmuebles = inmuebles.query("Tipo in ['Apartamento', 'Casa', 'Finca']")
fig = plt.figure(figsize=(12, 9))

sns.set_palette("rocket")

ax1 = plt.subplot2grid((2, 2), (1, 0))
ax1 = sns.histplot(data=three_inmuebles, x='Precio_Millon',element="step", 
                  kde=True, hue='Tipo')
ax1.set_xlim(50, 1500)
ax1.set_ylim(0, 10)
ax1.set_title("Comparacion de inmuebles(Finca)")
ax1.set_xlabel('Precio de los inmuebles', fontsize=13)
ax1.set_ylabel('Inmuebles en venta', fontsize=13)
ax1.xaxis.label.set_color('purple')
ax1.yaxis.label.set_color('purple')

ax2 = plt.subplot2grid((2, 2), (1, 1))
ax2 = sns.histplot(data=three_inmuebles, x='Precio_Millon', element="step",
                   kde=True, hue='Tipo')
ax2.set_xlim(50, 1500)
ax2.set_ylim(0, 100)
ax2.set_title("Comparacion de inmuebles(Casa)")
ax2.set_xlabel('Precio de los inmuebles', fontsize=13)
ax2.set_ylabel('Inmuebles en venta', fontsize=13)
ax2.xaxis.label.set_color('purple')
ax2.yaxis.label.set_color('purple')

ax3 = plt.subplot2grid((2, 2), (0, 0), colspan=2)
ax3 = sns.histplot(data=three_inmuebles, x='Precio_Millon', element="step",
                   kde=True, hue='Tipo')
ax3.set_xlim(50, 1500)
ax3.set_xlabel('Precio de los inmuebles', fontsize=13)
ax3.set_ylabel('Inmuebles en venta', fontsize=13)
ax3.xaxis.label.set_color('purple')
ax3.yaxis.label.set_color('purple')

plt.suptitle('Comparacion de inmuebles en Bogota')
fig.tight_layout(h_pad=3)
plt.show()
```
